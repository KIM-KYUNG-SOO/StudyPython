#### Python 폴더 관련 모듈
> 파일과 폴더의 차이: 확장자(파일의 성격을 규정하는 지침서)의 유무
1. 파일 폴더의 리스트화 및 추출
- 파일과 폴더를 구분
- 폴더내의 자료를 리스트화 할 수 있다.

```python
import os, glob
filter(os.path.isdir, glob.glob("./*")) # 현재폴더의 디렉토리와 모든 파일을 보고자 할때
dataList = glob.glob("./*") # 카멜표기법
# glob -> 사용자가 제시한 조건에 맞는 파일명을 리스트형식으로 반환
folderList = []
for x in dataList: # 객체할당
    print(x[2:])
    if '.' not in x [2:]:
        folderList.append(x)
folderList
```

- 위의 구문은 아래와 같이도 작성가능하다.

```python
# eee / jpg234.jpg
folderList = []
for x in dataList:
    if len(x[2:].split('.')) == 1:
        folderList.append(x)

folderList

# if x.split('.')[1] == 'jpg': # 참고 확장자에 대한 정보가 jpg라면
```

#### numpy 모듈을 이용한 연산

```python
import numpy as np
x = np.array([[22,3,4],[2,2,3]])
#[1번사람[x1,x2,x3],2번사람[x1,x2,x3]], x1을 몸무게/ x2를 나이/ x3를 수명량이라고 가정
y = np.array([100,150]) # 1번사람의 혈압이 100 / 2번사람의 혈압이 150일때
w = np.array([3,5,8]) # x1의 w값 3 / x2의 w값 5 / x3의 w값 8
np.sum(x)
np.array(x).sum()
# x.sum -> x는 np로 변환되지 않으면 리스트이므로, 합계가 되지 않는다.

y1 = x * w
y_hat = [] # 예측값
for i in y1:
    y_hat.append(i.sum())
print(y_hat)
type(y)

# NA: 입력되지 않은 값 -> 사칙연산이 안됨
# 즉 모든연산이 오류가 날 경우가 생긴다. NA로 들어오는 데이터를 어떻게 처리할 것이냐?
```

- 위의 for 구문은 아래와 같이 표현이 가능하다

```python
y_hat = [i.sum() for i in y1]
y_hat
```

#### for 컴프리헨션
> 첫 for 구문은 아래와 같이 사용이 가능하다.

```python
number = []
for x in range(10):
    number.append(x)
number
```

```python
number = [x for x in range(10)]
number
```
